#### Say you had a dataframe with two columns, sex and AQ. How would you boxplot male, female and total AQ?

```python
import numpy as np
import pandas as pd
import seaborn as sbn
import matplotlib.pylab as plt
%matplotlib inline
```

```python
sex = np.random.choice(['male','female'],40)
```

```python
sex
```

```python
AQ = np.random.randint(5,40,40)
```

```python
AQ
```

```python
data = pd.DataFrame({'sex':sex, 'AQ':AQ})
```

```python
data
```

### This way won't give you a total in the same graph

```python
data.groupby('sex').boxplot(return_type='axes')
```

```python
Create the columns you want to plot
```

```python
pd.DataFrame({'male' :  data.AQ[data.sex=='male'],
                      'female':data.AQ[data.sex=='female'],
                      'total':data.AQ})
```
